# Stage 01: Data Collection

Fetch historical F1 data (2014–2024) from the Ergast Developer API and save as raw CSV files.

**Data sources:**
- Race results (finishing positions, points, status)
- Qualifying results (Q1, Q2, Q3 times)
- Circuit information (name, country, coordinates)

```python
import sys
sys.path.append('..')

from src.data_pipeline import (
    fetch_race_results,
    fetch_qualifying_results,
    fetch_circuits,
    save_raw_data,
    get_data_summary
)
```

## 1.1 Fetch Race Results

```python
race_df = fetch_race_results(start_year=2014, end_year=2024)
print(f"Race results: {race_df.shape}")
race_df.head()
```

## 1.2 Fetch Qualifying Results

```python
qualifying_df = fetch_qualifying_results(start_year=2014, end_year=2024)
print(f"Qualifying results: {qualifying_df.shape}")
qualifying_df.head()
```

## 1.3 Fetch Circuit Information

```python
circuits_df = fetch_circuits()
print(f"Circuits: {circuits_df.shape}")
circuits_df.head()
```

## 1.4 Save Raw Data

```python
save_raw_data(race_df, qualifying_df, circuits_df)
```
